```python
import os
import boto3
ssm = boto3.client('ssm')
MODEL_ID = ssm.get_parameter(Name='/aim308/model-id')['Parameter']['Value']  # Claude Sonnet 4.5
REGION = ssm.get_parameter(Name='/aim308/region')['Parameter']['Value']
os.environ["AWS_REGION"] = REGION
os.environ["BYPASS_TOOL_CONSENT"] = os.getenv("BYPASS_TOOL_CONSENT", "true")
os.environ["STRANDS_TOOL_CONSOLE_MODE"] = "enabled"
os.makedirs("tools", exist_ok=True)
```

## The magic: `load_tools_from_directory=True`

```python
from strands import Agent
from strands_tools import shell, file_write
agent = Agent(
    model=MODEL_ID,
    tools=[shell, file_write],
    load_tools_from_directory=True,   # 👈 THE MAGIC
    system_prompt=(
        "You can create new tools by writing Python files to ./tools/. "
       "Decorate them with @tool from the strands package. "
       "They become available instantly."
    )
)
```

## Ask the agent to build a tool

```python
agent("create a tool called password_generator that makes random passwords, then use it to make a 20-char password")
```
